```python
df2['datetime'] = pd.to_datetime(df2['datetime'])
```

```python
seconds = ((df2.groupby(['visitor_id','session_id'])['datetime'].last()) - (df2.groupby(['visitor_id','session_id'])['datetime'].first())).apply(lambda x : x.seconds )
```

```python
seconds
```

```python
df3 = df2[(df2["session_id"].isin(list(data[1] for data in seconds[seconds <= 5].index))) &
         (df2["visitor_id"].isin(list(data[0] for data in seconds[seconds <= 5].index)))]
df3
```

```python
len(df3)
```

```python
output.update({"analysis 3":len(df3)})
output
```
